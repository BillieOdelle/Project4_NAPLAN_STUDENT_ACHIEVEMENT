```python
#import dependancies
import pandas as pd
from sqlalchemy import create_engine

database_path = pd.read_csv('Resources/Naplan_results_clean.csv')

#"postgresql+psycopg2://naplan_user:naplan_user@localhost/NAPLAN_DB"
```

```python
engine = create_engine('sqlite:///NAPLAN_SQLDB')
database_path.to_sql('naplan_results_clean', engine, index=False, if_exists='replace')
print(database_path.head())
```

```python
query = 'SELECT * FROM naplan_results_clean'
data = pd.read_sql_query(query, engine)

print(data.head(30))
```

```python
grouped_data = data.groupby('YEAR_LEVEL')
average_score_by_year = grouped_data['MEAN'].mean().round(0)

print(average_score_by_year)
```

```python
grouped_data = data.groupby(['YEAR_LEVEL', 'CALENDAR_YEAR'])
mean_score_by_year_level_calendar_year = grouped_data['MEAN'].mean().round(0)
print(mean_score_by_year_level_calendar_year)
```

```python
# Calculate descriptive statistics
print(data.describe())

# Group data by a specific column and calculate mean for other columns
print(data.groupby('DOMAIN').mean())

# Apply custom operations on a column
#data['new_column'] = data['column1'] + data['column2']
print(data.head())
```
