```python
new_df_Naplan = df_Naplan[df_Naplan.SUBGROUP == "All"]
new_df_Naplan
```

```python
df_1 = new_df_Naplan.groupby(['STATE','DOMAIN'])['ENROLLED_STUDENTS'].mean()
df_1
```

```python
df_2 = new_df_Naplan.groupby(['STATE'])['ENROLLED_STUDENTS'].mean()
df_3 = pd.DataFrame(df_2)
df_3.reset_index(inplace = True)
df_3 = df_3[df_3['STATE'] != 'AUS']
df_3 = df_3.sort_values(by='ENROLLED_STUDENTS', ascending=False)
df_3
```

```python
df_3['ENROLLED_STUDENTS'].sum()
```

```python
df_3.to_json('./data/enrolled_students.json',orient='records' )
```

```python
STATE = df_3['STATE']
ENROLLED_STUDENTS = df_3['ENROLLED_STUDENTS']
 
# Figure Size
fig = plt.figure(figsize =(10, 7))
 
# Horizontal Bar Plot
plt.bar(STATE, ENROLLED_STUDENTS)

plt.xlabel("State")
plt.ylabel("Enrolled Students")
plt.title("Average number of students attending NAPLAN test in each state from 2008 to 2022")
 
# Show Plot
plt.show()
```

```python
df_2 = new_df_Naplan.groupby(['STATE'])['MEAN'].mean()
df_3 = pd.DataFrame(df_2)
df_3.reset_index(inplace = True)
df_3 = df_3.sort_values(by='MEAN', ascending=False)
df_3
```

```python
df_3.to_json('./data/mean_scores.json',orient='records' )
```

```python
STATE = df_3['STATE']
MEAN_SCORE = df_3['MEAN']
 
# Figure Size
fig = plt.figure(figsize =(10, 7))
 
# Horizontal Bar Plot
plt.bar(STATE, MEAN_SCORE)

plt.xlabel("State")
plt.ylabel("Mean Score")
plt.title("Average score of NAPLAN test in each state from 2008 to 2022")
 
# Show Plot
plt.show()
```

```python
df_2 = new_df_Naplan.groupby(['STATE','CALENDAR_YEAR'])['MEAN'].mean()
df_3 = pd.DataFrame(df_2)
df_3.reset_index(inplace = True)
df_3 = df_3.sort_values(by=['STATE', 'CALENDAR_YEAR'])
df_3
```

```python
df_3.to_json('./data/mean_scores_by_state_and_year.json',orient='records' )
```

```python
Year = df_3['CALENDAR_YEAR'].unique()
Year
```

```python
States = df_3['STATE'].unique()
States
```

```python
for state in States:
    score = df_3[df_3['STATE'] == state]['MEAN']
    plt.plot(Year, score, label = state)

plt.legend()
# plt.savefig('Average_score_yearly_by_state.png')
plt.show()
```
